One of the questions I've been asking since I started working on the sort of experiments I did in my dissertation is how people acclimate to the task. When I would listen to participants as they stepped through the trials, I would notice them struggling to figure out how best to deal with the slightly convoluted rules of the task, particularly for the first few trials. Of course, any good psychologist will tell you that having participants practice on a strange task before engaging in the measured trials eliminates a lot of the variability in acclimating, and also of course, I took that advice. But the question of whether that was sufficient stuck with me, and today I wanted to sit down and see whether there was any pronounced difference between performance at the beginning of the task versus the end.

To do that, I wanted to take a look at how the onset to speak durations changed over the course of the task, from trial 1 to 100. If speakers were undergoing an acclimation process, the durations should decrease in aggregate in one of two ways: 1) a steady decline that would indicate they simply got faster with practice; 2) a steep decline over the first few trials, and then a steady average, indicating a brief acclimation period followed by "normal" behavior. The absence of an acclimation process (i.e. neither of these patterns - little to no average decrease) would suggest that participants used approximately the same cognitive mechanisms throughout, which I precisely what I wanted to measure. Honestly, it seems unlikely that participants would have experienced any notable difficulty with the task, especially after the practice trials. People talk all the time, in all sorts of strange situations, and this wasn't that much stranger than the norm. Still, in science it pays to check on unlikely possibilities once in a while.

Plus, it's good practice for creating complicated graphs!

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
from dissdata import importation

%matplotlib inline

# I find the constant opening and closing of my notebooks very annoying.
# This line ensures that the module gets reloaded every time I run something from it
%autoreload 2
exp = importation()
```

The goal with this set of graphics is to show onset durations plotted against trial numbers, y and x respectively. I also want to show a very basic linear plot of change over time (trial), so a simple linear regression (n.b. the regression setup I used in the actual analysis was significantly more complicated, but this is close enough for now). The graph below presents the basic idea - all of the points in the background, and the line overlaid to show how the average duration changes over the course of the experiment.

```python
x = exp.expcomp[0]['trial']
y = exp.expcomp[0]['onset_dur']
m, b = np.polyfit(x,y,1)

plt.plot(x,y,'.')
plt.plot(x,m*x + b, '-', color = 'r')
```

The panel is divided most usefully down the middle vertically - the left side has the experiments that had only-negative SOAs, and the right side has experiments with only-positive SOAs. Otherwise, all of the subplots depict the same thing, namely the durations plotted against trials, plus the line approximating the change over time. 

```python
exp.scatter_panel_soa(x='trial',y='onset_dur',xlab = 'Trial #', ylab = 'Duration (ms)', title = "Trial-level Onset Durations")
```

What's the answer? Do participants acclimate? It seems not - there are five slightly-negative slopes, two positive, and one almost entirely flat. One could conceivably argue that there is a slight downward trend, but it would be a hard sell. It is decidedly not the case that there was an acclimation period near the beginning (if there had been, those trials might need to be eliminated). A more detailed analysis of this question could break the graphs down into individual participants, to see if some acclimated while others didn't. On average there isn't a noticeable pattern, but some few might reveal improvement over time.
